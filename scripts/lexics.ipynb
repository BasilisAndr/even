```python
import os, re, pickle
from pprint import pprint
import numpy as np
import pandas as pd
```

```python
def get_wdchains(corp):
    res = []
    for doc in corp:
        for sent in corp[doc]['text']:
            if 'ps' in sent:
                if len(sent['ps'])!=len(sent['mb']):
                    print(doc)
                    continue
                s = [(sent['mb'][i][0], sent['ge'][i][0], sent['ps'][i][0]) for i in range(len(sent['ps']))]
                res += s
    return res

with open('Sebjan.pickle', 'rb') as f:
    seb = pickle.load(f)
with open('Kamchatka.pickle', 'rb') as f:
    kam = pickle.load(f)

from collections import Counter
seb_wds = Counter(get_wdchains(seb))
kam_wds = Counter(get_wdchains(kam))
```

```python
seb_r = [(*x, seb_wds[x]) for x in seb_wds if x[1].endswith('.R')]
kam_r = [(*x, kam_wds[x]) for x in kam_wds if x[1].endswith('.R')]

print(len(seb_r))
print(len(kam_r))

seb_pd = pd.DataFrame({
        'word': list(map(lambda x: x[0], seb_r)),
        'transl': list(map(lambda x: x[1], seb_r)),
        'pos': list(map(lambda x: x[2], seb_r)),
        's_count': list(map(lambda x: x[3], seb_r))
    })

kam_pd = pd.DataFrame({
        'word': list(map(lambda x: x[0], kam_r)),
        'transl': list(map(lambda x: x[1], kam_r)),
        'pos': list(map(lambda x: x[2], kam_r)),
        'k_count': list(map(lambda x: x[3], kam_r))
    })

tog = pd.merge(seb_pd, kam_pd, on=['word', 'transl', 'pos'], how='outer')
```
